```python
dataset = pd.DataFrame({2012:[medal_2012['Gender'][medal_2012['Gender']=='Male'].count(),medal_2012['Gender'][medal_2012['Gender']=='Female'].count(),event_2012['Men'].sum(),event_2012['Women'].sum()],
                       2016:[medal_2016['Gender'][medal_2016['Gender']=='Male'].count(),medal_2016['Gender'][medal_2016['Gender']=='Female'].count(),event_2016['Men'].sum(),event_2016['Women'].sum()],
                       2020:[medal_2020['Gender'][medal_2020['Gender']=='Male'].count(),medal_2020['Gender'][medal_2020['Gender']=='Female'].count(),event_2020['Men'].sum(),event_2020['Women'].sum()]})
```

```python
dataset = dataset.T
```

```python
dataset = dataset.rename(columns = {0:'Medals Won by Men',1:'Medals Won by Women',2:'Men Participants', 3:'Women Participants'})
```

```python
dataset['Men Medal Conversion Ratio'] = dataset['Medals Won by Men']/dataset['Men Participants']
dataset['Women Medal Conversion Ratio'] = dataset['Medals Won by Women']/dataset['Women Participants']
```
